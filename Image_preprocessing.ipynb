## Converting Images into train, validation and test folder

```python
import os
import shutil
import random

# Define the source directory containing class-based subfolders
source_directory = 'class_ based_structure'

# Define the parent directory where you have created train, validation, and test directories
parent_directory = 'train_validate_test'

# Define the split ratios (e.g., 70% train, 15% validation, 15% test)
train_ratio = 0.6
validation_ratio = 0.20
test_ratio = 0.20

# Iterate over the class-based subdirectories
for class_folder in os.listdir(source_directory):
    class_path = os.path.join(source_directory, class_folder)

    # Create corresponding subdirectories in the parent directory
    train_dir = os.path.join(parent_directory, 'train', class_folder)
    validation_dir = os.path.join(parent_directory, 'validation', class_folder)
    test_dir = os.path.join(parent_directory, 'test', class_folder)

    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(validation_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # List all image files in the class folder
    image_files = os.listdir(class_path)

    # Shuffle the image files randomly
    random.shuffle(image_files)

    # Determine the number of images for each split
    total_images = len(image_files)
    num_train = int(total_images * train_ratio)
    num_validation = int(total_images * validation_ratio)

    # Distribute the images to train, validation, and test folders
    train_images = image_files[:num_train]
    validation_images = image_files[num_train:num_train + num_validation]
    test_images = image_files[num_train + num_validation:]

    # Move the image files to their respective folders
    for image in train_images:
        src = os.path.join(class_path, image)
        dst = os.path.join(train_dir, image)
        shutil.copy(src, dst)

    for image in validation_images:
        src = os.path.join(class_path, image)
        dst = os.path.join(validation_dir, image)
        shutil.copy(src, dst)

    for image in test_images:
        src = os.path.join(class_path, image)
        dst = os.path.join(test_dir, image)
        shutil.copy(src, dst)
```
